## Inicialmente se realizará una extracción de todos los datos de los archivos a analizar y una modelación de estos para un mejor análisis 
##### Con la función glob cogemos todos los csvs y los leemos con la libreria pandas para crear los data frames. Para mejorar el analisis de los datos se han creado los campos dia mes y año a apartir del nombre del csv.

```python
import pandas as pd
import numpy as np
import glob
import geopandas as gpd
from pylab import *
```

```python
list_dfs = []
meteo_dfs = []

csv_files = glob.glob('CSV/*')
for file in csv_files:
    data = pd.read_csv(file, on_bad_lines='skip', delimiter=';')
    data["dia"] = int(file.split('-')[2].split('.')[0])
    data["mes"] = int(file.split('-')[1])
    data["año"] = int(file.split('-')[0].split('Aemet')[1])
    list_dfs.append(data)
  
meteo_dfs = pd.concat(list_dfs)
meteo_dfs = meteo_dfs.rename(columns={'Temperatura máxima (ºC)': 'tmax', 'Temperatura mínima (ºC)':'tmin', 'Temperatura media (ºC)':'tmedia', 'Racha (km/h)':'racha', 'Velocidad máxima (km/h)':'vmax', 'Precipitación 00-24h (mm)':'precip24', 'Precipitación 00-06h (mm)':'precip06', 'Precipitación 06-12h (mm)':'precip612', 'Precipitación 12-18h (mm)':'precip1218'})
```

## Visualización de datos básicos
### Temperatura máxima y mínima de los años 2018, 2019, 2021 y 2022
#### Obtención de datos
##### Se obtiene la máxima y mínima temperatura del dataframe iterado por cada año desde 2018 hasta 2022, para obtener el dato de cada uno de estos años.

```python
k = 2018
df = []
tmax = []
tmin = []
año = []
while k<= 2022:
    df = meteo_dfs[meteo_dfs["año"]==k]
    max_temp = df["tmax"].str.split().str[0].astype(float)
    tmax.append(max(max_temp))
    min_temp = df["tmin"].str.split().str[0].astype(float)
    tmin.append(min(min_temp))
    año.append(k) 
    k=k+1
```

#### Exposición de los datos en el gráfico

```python
plt.figure(figsize=(20,10))
plt.plot(tmax, label="Máximas", marker=".", color="red")
plt.plot(tmin, label="Mínimas", marker=".", color="blue")
plt.xlabel("Año")
plt.ylabel("Temperatura ºC")
plt.title(" Temperatura máxima y mínima en España")
plt.legend(loc="right")
indice = np.arange(len(año))
plt.xticks(indice, año)
plt.grid(True)
plt.show()
plt.close()
```

### Los 10 municipios con mayores rachas de 2021
#### Obtención de datos
##### Se crea una lista de municipios, luego se itera el dataframe del año 2021 sobre los municipios de la lista y se extrae la racha máxima de cada municipio. Posteriormente, se gurada en un diccionario siendo la clave el municipio y el valor la racha máxima obtenida. Este diccionario se ordena y se obtienen los 10 municipios primeros que son los que tienen las rachas más fuertes.

```python
list_mun = []
df = meteo_dfs[meteo_dfs["año"]==2021]
df_i = []
racha_prov = {}
top10 = {}

for i in df.loc[:,"Estación"]:
    if i not in list_mun:
        list_mun.append(i)

for mun in list_mun:
    df_i = df[df.loc[:,"Estación"]==mun]
    value = df_i["racha"].str.split().str[0].astype(float).fillna(0)
    value = max(value)
    racha_prov[mun]=value

racha_ord = dict(sorted(racha_prov.items(), key=lambda item: item[1], reverse=True)[0 : 10]) 
```

#### Exposición de los datos en el gráfico

```python
plt.figure(figsize=(25,10))
ax = plt.axes()
plt.bar(racha_ord.keys(),racha_ord.values(), align ="center", width=0.6, color=['#9afcb3','#deb8ff','#f2a677','#82e2fa','#ffb3f5','#f5e6a6','#ff9999','#7091fa','#d7fa70','#c7abab'], edgecolor="none")
ax.yaxis.grid(True, linestyle=':')
plt.xlabel("Municipio")
plt.ylabel("Racha Km/h")
plt.title("Los 10 municipios con las rachas más fuertes de España en 2021")
indice = np.arange(10)
plt.xticks(indice, racha_ord.keys())
ax.spines['bottom'].set_linewidth(2)
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('none')
plt.show()
plt.close()
```

## Visualización de diferentes tipos de datos 
### Climograma en el que se muestra la temperatura media y la velocidad máxima del viento de cada mes en el año 2020 en el municipio Rasquera de Tarragona.
#### Obtención de datos
##### Se obtiene la temperatura media, la velocidad máxima y la racha media del viento del dataframe filtrado por los meses del año y el municipio Rasquera.

```python
i=1
t_med = []
v_max = []
racha_med = []
df = []
df = meteo_dfs[meteo_dfs["año"]==2020]
df_i = df[df["Estación"]=='Rasquera']

while i<=12:
    df_a = df_i[df_i["mes"]==i]
    value = df_a.loc[:,"tmedia"].mean()
    t_med.append(value)
    value = df_a.loc[:,"vmax"].str.split().str[0].astype(float).max()
    v_max.append(value)
    value = df_a.loc[:,"racha"].str.split().str[0].astype(float).mean()
    racha_med.append(value)
    i=i+1
```

#### Exposición de los datos en el gráfico

```python
meses = ['Enero','Febrero','Marzo','Abril','Mayo','Junio','Julio','Agosto','Septiembre','Octubre','Noviembre','Diciembre']
plt.figure(figsize=(25,10))
ax = plt.axes()

ax.set_xlabel("Mes")
ax.set_ylabel("Temperatura ºC")
plt.bar(meses,t_med, align ="center", width=0.4, facecolor='#9999ff')
ax2 = ax.twinx()
ax2.set_ylabel('Km/h')
plt.plot(v_max, label="Velocidad Máxima", marker=".", color="black")
plt.plot(racha_med, label="Racha media", marker=".", color="grey", linestyle='--')


plt.grid(True, color='#999fff')
plt.legend(loc="upper right")

plt.show()
plt.close()
```

## Visualización compleja
### Mapa de precipitaciones para comparar las 4 estaciones en España en 2021.
#### Obtención de datos

##### Primero haremos una lista de todas las provincias de españa para obtener la media de lluvias de cada una de ellas en cada estación

```python
list_df = []
df_estacion = []
list_prov = []
df = meteo_dfs[meteo_dfs["año"]==2021]

for i in df["Provincia"]:
    if i not in list_prov:
        list_prov.append(i)
```

##### Creación del datadrame de invierno
##### Crearemos un dataframe nuevo que vaya desde el 22 de diciembre hasta el 22 de marzo para crear un mapa sobre este nuevo dataframe.

```python
df_i = df[df["mes"]==12]
list_df.append(df_i[df_i["dia"]>=22])
list_df.append(df[df["mes"]<3])
df_i = df[df["mes"]==3]
list_df.append(df_i[df_i["dia"]<=19])
df_estacion = pd.concat(list_df)

inv = {}
for i in list_prov:
    df_estacion_i = df_estacion[df_estacion['Provincia']==i]
    inv[i] = df_estacion_i.loc[:,'precip24'].astype(float).mean()

df_precipitaciones = pd.DataFrame({"provincia": inv.keys(), "Precip 24": inv.values()})
```

##### Creación del mapa de invierno

```python
shapefile = 'provincias-espanolas.geojson'
spain_gdf = gpd.read_file(shapefile)
spain_invierno = spain_gdf.merge(df_precipitaciones, on="provincia")
```

```python
fig, ax = plt.subplots(figsize=(20, 15))
ax.set_title("Media de lluvias en invierno de 2021"+"\n",fontsize=30,color='#4873ab')

from mpl_toolkits.axes_grid1 import make_axes_locatable
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="2%", pad=0.1)

spain_invierno.plot(column = 'Precip 24', cmap='Blues', ax=ax, legend=True, cax=cax, zorder=5)
plt.show()
plt.close()
```
